```python
import sys, os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime, timedelta

FMT = '%Y-%m-%d'
get_ts = lambda dt: datetime.strftime(dt, FMT)
get_dt = lambda ts: datetime.strptime(ts, FMT)

DAYS = 366 + 63
TSFIRST = '2019-12-01'
AMAX = np.exp(DAYS / 31)
TSLAST = get_ts(get_dt(TSFIRST) + timedelta(DAYS - 1))

dat = [
  ['2020-01-28', 106],
  ['2020-01-29', 132],
  ['2020-01-30', 170],
  ['2020-01-31', 213],
  ['2020-02-01', 259],
  ['2020-02-02', 304],
  ['2020-02-03', 362],
  ['2020-02-04', 426],
  ['2020-02-05', 491],
  ['2020-02-06', 563],
  ['2020-02-07', 634]]

def new_corona(df):
  fig, ax = plt.subplots(nrows=1, ncols=1)
  print(df.iloc[55:70,:])
  df.dropna().plot(ax=ax, style='b.-') # , kind='bar')
  dfr = df.iloc[:,:]
  dfr.ix[TSFIRST] = (0.0, )
  dfr.ix[TSLAST] = (AMAX, )
  print(dfr.iloc[[0],:])
  print(dfr.iloc[[-1],:])
  dfr.interpolate(method='quadratic').plot(ax=ax, style='r-') # cubic, index
  ax.set_xlim([TSFIRST, TSLAST])
  ax.set_ylim([0, AMAX])
  ax.legend(loc='best')
  plt.show()

idx = np.zeros(DAYS)
ts = pd.Series(idx, index=pd.date_range(TSFIRST, periods=DAYS))
val = np.full((DAYS, 1), np.nan)
df = pd.DataFrame(val, index=pd.date_range(TSFIRST, periods=DAYS), columns=['val'])
#df = df.set_index('date')
for d in dat:
  ymd, v = d[0], float(d[1])
  df.ix[ymd] = (v, )
new_corona(df)
```
